```python
#Gets a video input path, an output path for the folder to contain the images and every count seconds.
#If the count is 0, the default value is every possible frame
def getFramesFromVideo( videoInputPath, outputFolderPath, count ):
    import cv2
    vidcap = cv2.VideoCapture(videoInputPath) #example: "D:\\project files\\information extraction\\mp4\\canon750D.mp4"
    
    if (count != 0):
        vidcap.set(cv2.CAP_PROP_POS_MSEC,(count*1000))    # every 'count' seconds
    
    success,image = vidcap.read()
    count = 0
    success = True
    
    while success:
      cv2.imwrite((outputFolderPath + "\\frame%d.jpg") % count, image)     # save frame as JPEG file, example: "D:\\project files\\information extraction\\imgExtraction\\Canon750D\\frame%d.jpg" % count, image
      success,image = vidcap.read()
      #print 'Read a new frame: ', success
      count += 1
    
   #return
```
